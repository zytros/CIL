```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_multiple_arrays(arrays, labels=None, title='Multiple Arrays Plot', xlabel='X-axis', ylabel='Y-axis'):
    """
    Plots multiple arrays on the same graph.
    
    Parameters:
    *arrays: Variable length argument list of arrays to plot.
    labels: List of labels for each array (optional).
    title: Title of the plot (optional).
    xlabel: Label for the X-axis (optional).
    ylabel: Label for the Y-axis (optional).
    """
    for i in range(arrays.shape[0]):
        label = labels[i] if labels and i < len(labels) else f'Array {i+1}'
        plt.plot(array[i], label=label)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
def plot_array(array, label='Array', title='Array Plot', xlabel='X-axis', ylabel='Y-axis'):
    """
    Plots a single array.
    
    Parameters:
    array: Array to plot.
    label: Label for the array (optional).
    title: Title of the plot (optional).
    xlabel: Label for the X-axis (optional).
    ylabel: Label for the Y-axis (optional).
    """
    plt.plot(array, label=label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```
